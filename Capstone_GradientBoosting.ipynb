```python
import pandas as pd

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV, LeaveOneOut

# Load data  #GBR #GridSearch #LOO #letswork
df = pd.read_csv("../Capstone/Capstone Dataset (New).csv")

# Split features/target  #cleanup
X = df.drop(
    ["Array type", "Predicted angular insertion depth (deg)",
     "Actual angular insertion depth (deg)"],
    axis=1
)
y = df["Actual angular insertion depth (deg)"]

# Model  #treeboost
gbr = GradientBoostingRegressor(random_state=42)

# Search space  #knobs
param_grid = {
    "n_estimators": [100, 200, 300],
    "learning_rate": [0.01, 0.05, 0.1, 0.2],
    "max_depth": [2, 3, 4, 5],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 5],
    "subsample": [1.0, 0.7],          # 0.7 = stochastic gradient boosting  #speedvsvariance
    "max_features": [None, "sqrt", "log2"]
}

# Leave-One-Out CV  #oneoutatatime #tightsplit
cv = LeaveOneOut()

# Grid search  #turnthecrank
gs = GridSearchCV(
    estimator=gbr,
    param_grid=param_grid,
    scoring="neg_mean_absolute_error",
    cv=cv,
    n_jobs=-1,
    verbose=1,
    refit=True
)

# Run it  #receipts
gs.fit(X, y)

print("Best params:", gs.best_params_)   # #facts
print("CV best MAE:", -gs.best_score_)   # #numbersdontlie
```
